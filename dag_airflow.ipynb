```python
dag_code = """
from airflow import DAG
from airflow.operators.bash import BashOperator
from datetime import datetime

with DAG(
    "etl_spark_pipeline",
    start_date=datetime(2024, 1, 1),
    schedule_interval="@daily",
    catchup=False,
    default_args={
        'retries': 1,
    }
) as dag:

    extract = BashOperator(
        task_id="extract_task",
        bash_command="echo 'Lancement Extract Bronze'"
    )

    transform = BashOperator(
        task_id="transform_task",
        bash_command="echo 'Lancement Transform Silver'"
    )

    aggregate = BashOperator(
        task_id="aggregate_task",
        bash_command="echo 'Lancement Aggregate Gold'"
    )

    extract >> transform >> aggregate
"""

# Sauvegarder le fichier DAG
with open('/content/drive/MyDrive/ETL_projet/dag_etl.py', 'w') as f:
    f.write(dag_code)

print("вњ… Fichier DAG sauvegardГ© !")
print(dag_code)
```

```python
import subprocess
result = subprocess.run(
    ['python', '/content/drive/MyDrive/ETL_projet/dag_etl.py'],
    capture_output=True, text=True
)
print("Sortie :", result.stdout)
print("Erreurs :", result.stderr)
if result.returncode == 0:
    print("вњ… DAG valide вЂ” aucune erreur de syntaxe !")
else:
    print("вќЊ Erreur dans le DAG")
```
