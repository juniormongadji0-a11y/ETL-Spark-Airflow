```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("ETL_Extract") \
    .getOrCreate()

print("✅ Spark démarré :", spark.version)
```

```python
import os

for root, dirs, files in os.walk('/content/drive/MyDrive'):
    for file in files:
        if 'retail' in file.lower() or 'online' in file.lower():
            print(os.path.join(root, file))
```

```python
import pandas as pd

# Lecture du fichier CSV
df_pandas = pd.read_csv(
    '/content/drive/MyDrive/ETL_projet/archive (1)/online_retail_II.csv',
    encoding='utf-8'
)

print("✅ Fichier lu :", df_pandas.shape[0], "lignes,", df_pandas.shape[1], "colonnes")
print(df_pandas.head(3))
```

```python
# Conversion pandas → Spark
df_spark = spark.createDataFrame(df_pandas)

# Sauvegarde en Parquet (couche Bronze)
df_spark.write.mode("overwrite").parquet(
    '/content/drive/MyDrive/ETL_projet/bronze/'
)

print("✅ Bronze sauvegardé !")
print("Nombre de lignes :", df_spark.count())
```
